```python
# 製作ChatGPT自訂函式
def askchatgpt(Question):
    response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {
        "role": "user",
        "content": Question
        },

        {
        "role": "assistant",
        "content": ""
        },

    ],
    #prompt="請問Python有哪幾種基本的資料型態？",
    temperature=0,
    max_tokens=500,
    top_p=1,
    frequency_penalty=0.5,
    presence_penalty=0
    )


    # 顯示答案
    choices = getattr(response, 'choices', [])

    # 假设 choices 列表中的每个元素都是 Choice 类的实例
    # 并且 Choice 类有一个名为 message 的属性，是一个对象
    # 这个 message 对象又有一个名为 content 的属性

    # 尝试获取第一个 choice 的 message 的 content
    # 返回答案
    if choices:
        first_choice = choices[0]
        message = getattr(first_choice, 'message', None)
        if message:
            content = getattr(message, 'content', '默认值，如果没有找到 content')
            #print(content)
            return content

        else:
            #print("没有找到 message")
            return "無資料"
    else:
        #print("choices 列表为空")
        return "無資料"
```

```python
Question = input("請輸入您的問題：")
print(askchatgpt(Question))
```

```python
# 完整的ChatGPT簡易聊天程式
from openai import OpenAI
import os
from dotenv import load_dotenv #匯入環境變數套件

load_dotenv()    #載入環境變數

# 製作有使用介面的聊天機器人：Gradio
# pip install gradio

#create interface object
import gradio as gr

#設定環境變數API_Key
client = OpenAI(
    # This is the default and can be omitted
    api_key = os.getenv("API_KEY"),
)

# 製作ChatGPT自訂函式
def askchatgpt(Question):
    response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {
        "role": "user",
        "content": Question
        },

        {
        "role": "assistant",
        "content": ""
        },

    ],
    #prompt="請問Python有哪幾種基本的資料型態？",
    temperature=0,
    max_tokens=500,
    top_p=1,
    frequency_penalty=0.5,
    presence_penalty=0
    )

    # 取得內容
    choices = getattr(response, 'choices', [])

    # 返回答案
    if choices:
        first_choice = choices[0]
        message = getattr(first_choice, 'message', None)
        if message:
            content = getattr(message, 'content', '默认值，如果没有找到 content')
            #print(content)
            return content

        else:
            #print("没有找到 message")
            return "無資料"
    else:
        #print("choices 列表为空")
        return "無資料"

#主程式
gui = gr.Interface(fn=askchatgpt, #callable function
                   inputs="text", #input format
                   outputs="text") #output format
#display the interface
gui.launch(share=True)
```
